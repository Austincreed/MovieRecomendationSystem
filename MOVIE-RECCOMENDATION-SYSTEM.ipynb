```python
#helper function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): #ast.literal_eval will covert string of list to list
        L.append(i['name'])
    return L
```

```python
#applying the effect in function in genres
movies['genres'].apply(convert) 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
#taking only 1st 3 names of actors from cast
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj): #converting string to list
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
            
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#this is used to fetch the name of the director from the large list of dictionaries of crew
def fetch_director(obj): 
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#converting the string in overview to list so that we can concatentate with other lists
movies['overview'].apply(lambda x:x.split())
```

```python
#changing string to list using .split()
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#NOW WE HAVE TO DO TRANSFORMATION IN EVERY ITEMS
#LIKE we have to remove all space 'sam worthington' -> 'samworthingtom'
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
# replace all " "spaces with "" no spaces
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#now creating a new tag to concatenate some columns|
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#helper function
def stem(text): #steming
    y = []
    for i in text.split(): #string to list
        y.append(ps.stem(i)) #stemming and append to y list 
    
    return " ".join(y) #list to string
    
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
#count vectorizer returns a scikit spars matrix
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
#storing the sparse skin kit matrix into vectors variable
vectors = cv.fit_transform(new_df['tags']).toarray() 
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
#this will measure the cosine distance between movies
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape #here we are calculating distance between one movie and 4806 movies and this will be done for all this is done so 4806 movies and 4806 distances for each(4806, 4806)
```

```python
similarity = cosine_similarity(vectors) #this store the array of distances
```

```python
#indexing the distances in the array
list(enumerate(similarity[0])) #list of tuples in result
```

```python
sorted(list(enumerate(similarity[0]))) #according to 1st index but we need according to 2nd index
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1]) #sorted according to distance not according to index in tuple
```

```python
# working is take the similarity[index of the movie] then we will get the array of idtances betwwn all movies from this movie then we will sort the list and will return 1st 5 similar movies from our list

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #sorted movie list ka 1st 5 movies
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
              
```

```python
recommend('Batman')
```
